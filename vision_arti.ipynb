```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de transformación lineal
def transformacion_lineal(r, a, b):
    return a * r + b

# Definir la función de transformación logarítmica
def transformacion_logaritmica(r, c):
    return c * np.log(1 + r)

# Definir la función de transformación de potencia
def transformacion_potencia(r, gamma):
    return np.power(r, gamma)


# Definir la función de transformación de potencia
def transformacion_negativa(r):
    return 1 - r

# Generar un rango de valores de intensidad de entrada (r)
r = np.linspace(0, 1, 100)

# Parámetros para las transformaciones
a, b = 1.5, 0.3  # para la transformación lineal
c = 1            # para la transformación logarítmica
gamma = 2        # para la transformación de potencia

# Calcular los valores transformados (s) para cada tipo de transformación
s_lineal = transformacion_lineal(r, a, b)
s_logaritmica = transformacion_logaritmica(r, c)
s_potencia = transformacion_potencia(r, gamma)
s_negativa = transformacion_negativa(r)

# Graficar las transformaciones de intensidad
plt.figure(figsize=(8, 4))

# Transformación lineal
plt.plot(r, s_lineal, 'r-', label='Transformación Lineal')
plt.title('Transformación Lineal')
plt.xlabel('Valor de entrada (r)')
plt.ylabel('Valor de salida (s)')
plt.legend()
# Guardar las gráficas como imágenes en el directorio local
plt.savefig('transformacionesLineal.png')

# Graficar las transformaciones de intensidad
plt.figure(figsize=(8, 4))
# Transformación logarítmica
plt.plot(r, s_logaritmica, 'g-', label='Transformación Logarítmica')
plt.title('Transformación Logarítmica')
plt.xlabel('Valor de entrada (r)')
plt.ylabel('Valor de salida (s)')
plt.legend()
# Guardar las gráficas como imágenes en el directorio local
plt.savefig('transformacionesLogarítmica.png')

# Graficar las transformaciones de intensidad
plt.figure(figsize=(8, 4))

# Transformación de potencia
plt.plot(r, s_potencia, 'b-', label='Transformación de Potencia')
plt.title('Transformación de Potencia')
plt.xlabel('Valor de entrada (r)')
plt.ylabel('Valor de salida (s)')
plt.legend()
# Guardar las gráficas como imágenes en el directorio local
plt.savefig('transformacionesPotencia.png')
plt.tight_layout()
plt.show()



# Graficar las transformaciones de intensidad
plt.figure(figsize=(8, 4))
# Transformación logarítmica
plt.plot(r, s_negativa, 'y-', label='Transformación Negativo')
plt.title('Transformación Negativo')
plt.xlabel('Valor de entrada (r)')
plt.ylabel('Valor de salida (s)')
plt.legend()
# Guardar las gráficas como imágenes en el directorio local
plt.savefig('transformacionesNegativo.png')
```

```python
# Cargar la imagen
imagen = mpimg.imread('5996.png')

# Convertir la imagen a escala de grises si es en color
if len(imagen.shape) > 2:
    imagen_gris = np.dot(imagen[...,:3], [0.2989, 0.5870, 0.1140])
else:
    imagen_gris = imagen
# Obtener el histograma de la imagen
hist, bins = np.histogram(imagen_gris.flatten(), bins=256, range=(0,1))

# Calcular la función de distribución acumulada (CDF)
cdf = hist.cumsum()

# Normalizar la CDF para ajustarla al rango [0, 1]
cdf_normalized = cdf / float(cdf.max())

# Mapear los valores de intensidad originales a los nuevos valores normalizados
imagen_normalizada = np.interp(imagen_gris.flatten(), bins[:-1], cdf_normalized)

# Reformar la imagen con los nuevos valores normalizados
imagen_normalizada = imagen_normalizada.reshape(imagen_gris.shape)

# Calcular el histograma de la imagen original manualmente
hist_original, bins_original = np.histogram(imagen_gris.flatten(), bins=256, range=(0,1))

# Normalizar el histograma de la imagen original
hist_normalizado, bins_normalizado = np.histogram(imagen_normalizada.flatten(), bins=256, range=(0,1))
hist_normalizado = hist_normalizado / np.sum(hist_normalizado)

# Mostrar la imagen original y la imagen con histograma normalizado
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(imagen_gris, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(imagen_normalizada, cmap='gray')
plt.title('Histograma Ecualizado')
plt.axis('off')
mpimg.imsave("histo5996.png", imagen_normalizada, cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2


# Cargar la imagen
imagen = mpimg.imread('5996.png')

# Convertir la imagen a escala de grises si es en color
if len(imagen.shape) > 2:
    imagen_gris = np.dot(imagen[...,:3], [0.2989, 0.5870, 0.1140])
else:
    imagen_gris = imagen


def modificacion_logaritmica(imagen, c, sigma):
    # Aplicar la transformación logarítmica a cada píxel de la imagen
    imagen_modificada = c * np.log(1 + (np.exp(sigma) - 1) * imagen)

    # Asegurarse de que los valores estén en el rango [0, 1]
    imagen_modificada = np.clip(imagen_modificada, 0, 1)

    return imagen_modificada



# Parámetros para la modificación logarítmica
c = 5  # Constante de ajuste
sigma = 0.5  # Factor de contraste

# Aplicar la transformación a cada píxel de la imagen
imagen_transformada = modificacion_logaritmica(imagen_gris,c, sigma)

# Mostrar la imagen original y la imagen transformada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen_gris, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')


plt.subplot(1, 2, 2)
plt.imshow(imagen_transformada, cmap='gray')
plt.title('Imagen Transformada logartimica')
plt.axis('off')

plt.show()



mpimg.imsave("COLineal5996.png", imagen_transformada, cmap='gray')
# Cargar la imagen

# Función para la corrección de gamma
def correccion_gamma(imagen, gamma):
    # Aplicar la corrección de gamma a cada píxel de la imagen
    imagen_corregida = np.power(imagen, gamma)

    # Asegurarse de que los valores estén en el rango [0, 1]
    imagen_corregida = np.clip(imagen_corregida, 0, 1)

    return imagen_corregida

# Parámetro gamma para la corrección de gamma
gamma = 1

# Aplicar la corrección de gamma
imagen_corregida = correccion_gamma(imagen_transformada, gamma)

# Mostrar la imagen original y la imagen con corrección de gamma
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen_gris, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagen_corregida, cmap='gray')
plt.title('Corrección de Gamma')
plt.axis('off')

plt.show()
mpimg.imsave("COgamma5996.png", imagen_corregida, cmap='gray')


# Obtener el histograma de la imagen
hist, bins = np.histogram(imagen_corregida.flatten(), bins=256, range=(0,1))

# Calcular la función de distribución acumulada (CDF)
cdf = hist.cumsum()

# Normalizar la CDF para ajustarla al rango [0, 1]
cdf_normalized = cdf / float(cdf.max())

# Mapear los valores de intensidad originales a los nuevos valores normalizados
imagen_normalizada = np.interp(imagen_corregida.flatten(), bins[:-1], cdf_normalized)

# Reformar la imagen con los nuevos valores normalizados
imagen_normalizada = imagen_normalizada.reshape(imagen_corregida.shape)


# Convertir la imagen a escala de grises si es necesario
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Crear un objeto CLAHE con un límite de contraste específico
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# Aplicar la ecualización adaptativa del histograma con límite
imagen_ecualizada = clahe.apply(imagen_corregida)

# Convertir la imagen a escala de grises si es necesario
imagen_gris = cv2.cvtColor(imagen_corregida, cv2.COLOR_BGR2GRAY)

# Aplicar la umbralización adaptativa
umbral_adaptativo = cv2.adaptiveThreshold(imagen_gris, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

# Opcional: Aplicar un filtro de mediana para suavizar la imagen binarizada
umbral_adaptativo_suavizado = cv2.medianBlur(umbral_adaptativo, 3)


# Mostrar la imagen original y la imagen con histograma normalizado
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(imagen_ecualizada, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 3, 1)
plt.imshow(umbral_adaptativo, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')


plt.subplot(1, 3, 2)
plt.imshow(umbral_adaptativo_suavizado, cmap='gray')
plt.title('Histograma Ecualizado')
plt.axis('off')
mpimg.imsave("COhisto5996.png", imagen_normalizada, cmap='gray')


# Función para realizar multiplicación por escalar en cada píxel de la imagen
def multiplicacion_escalar(imagen, escala):
    # Multiplicar la imagen por el valor de escala
    resultado = imagen * escala
    return resultado


# Valor de escala para multiplicar la imagen
escala = 300 # Cambia este valor según lo que desees multiplicar

# Aplicar la multiplicación por escalar a la imagen
imagen_resultante = multiplicacion_escalar(imagen_normalizada, escala)

# Asegurarse de que los valores estén en el rango [0, 255]
imagen_resultante = np.clip(imagen_resultante, 0, 255).astype(np.uint8)


# Mostrar la imagen original y la imagen resultante en blanco y negro
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen_normalizada, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagen_resultante, cmap='gray')
plt.title('Imagen Resultante')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

# Cargar la imagen
imagen = mpimg.imread('4391.png')

# Convertir la imagen a escala de grises si es en color
if len(imagen.shape) > 2:
    imagen_gris = np.dot(imagen[...,:3], [0.2989, 0.5870, 0.1140])
else:
    imagen_gris = imagen

# Ajustes para mantener legible el texto del letrero de neón y mejorar la visibilidad en partes oscuras
c = 5  # Reducir el factor de contraste para la transformación logarítmica
sigma = 0.5  # Reducir el factor de contraste para la transformación logarítmica
gamma = 1.2  # Aumentar el valor de gamma para la corrección de gamma
clipLimit = 6.0  # Aumentar el límite de contraste para la ecualización adaptativa del histograma con límite
tileGridSize = (8, 8)  # Aumentar el tamaño de la cuadrícula para la ecualización adaptativa del histograma con límite

# Función para aplicar la modificación logarítmica a la imagen
def modificacion_logaritmica(imagen, c, sigma):
    imagen_modificada = c * np.log(1 + (np.exp(sigma) - 1) * imagen)
    imagen_modificada = np.clip(imagen_modificada, 0, 1)
    return imagen_modificada

# Aplicar la modificación logarítmica
imagen_transformada = modificacion_logaritmica(imagen_gris, c, sigma)

# Aplicar la corrección de gamma
imagen_corregida = np.power(imagen_transformada, gamma)
imagen_corregida = np.clip(imagen_corregida, 0, 1)

# Crear el objeto CLAHE para la ecualización adaptativa del histograma con límite
clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize)
imagen_ecualizada = clahe.apply((imagen_corregida * 255).astype(np.uint8))


# Mostrar las imágenes
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(imagen_corregida, cmap='gray')
plt.title('Imagen Corregida')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(imagen_ecualizada, cmap='gray')
plt.title('Imagen Ecualizada')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(imagen_gris, cmap='gray')
plt.title('Imagen original')
plt.axis('off')


plt.show()

mpimg.imsave("COhistolimit4391.png", imagen_ecualizada, cmap='gray')
```
